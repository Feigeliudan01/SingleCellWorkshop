```python
# %load /home/dan/.jupyter/firstcell.py
%load_ext autoreload
%autoreload 2

import sys
import os
import time

%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns

font = {'size'   : 14}
mpl.rc('font', **font)

import numpy as np
import pandas as pd

from scipy import stats
import sklearn
from sklearn import cluster
import scprep
```

```python
np.random.seed(42)
data = [np.random.normal(size=(100,2)),
        np.random.normal([0,5],size=(100,2)),
        np.random.normal([7,0],size=(100,2))]

data = np.concatenate(data,axis=0)
```

```python
labels = np.zeros(300, int)
labels[100:200] =1
labels[200:] = 2
```

```python
fig, ax = plt.subplots(1, figsize=(4,4))

scprep.plot.scatter2d(data, title='Data', c=labels,
                      xlabel=r'$x_1$',
                      ylabel=r'$x_2$',
                      ax=ax)

fig.tight_layout()
fig.savefig('../img/DiffExp.data.png', dpi=300)
```

```python
fig, ax = plt.subplots(1, figsize=(5,3))

for l in np.unique(labels):
    scprep.plot.histogram(data[:,0][labels == l], range=(-3,11), bins=20, alpha = 0.33, label=l,
                    xlabel=r'$x_1$',
                      ylabel='frequency',ax=ax)
    
ax.legend()


fig.tight_layout()
fig.savefig('../img/DiffExp.distribution.x1.png', dpi=300)
```

```python
fig, ax = plt.subplots(1, figsize=(5,3))

for l in np.unique(labels):
    scprep.plot.histogram(data[:,1][labels == l], range=(-3,11), bins=20, alpha = 0.33, label=l,
                    xlabel=r'$x_2$',
                      ylabel='frequency',ax=ax)
    
ax.legend()


fig.tight_layout()
fig.savefig('../img/DiffExp.distribution.x2.png', dpi=300)
```

```python
data3 = np.zeros(600)
data3[0:200] = np.random.binomial(2, 0.5, 200)
data3[200:400] = np.random.binomial(18, 0.5, 200)
data3[400:] = np.random.binomial(12, 0.5, 200)
```

```python
labels = np.zeros(600, int)
labels[200:400] =1
labels[400:] = 2
```

```python
fig, axes = plt.subplots(3, figsize=(5,9))

for i, l in enumerate(np.unique(labels)):
    
    scprep.plot.histogram(data3[labels == l], range=(0,15), bins=15, rwidth=0.5, 
                      alpha = 0.33, label=l,
                      xlabel=r'$x_2$',
                      ylabel='frequency',ax=axes[i])
    
ax.legend()


fig.tight_layout()
fig.savefig('../img/DiffExp.distribution.x3.png', dpi=300)
```
