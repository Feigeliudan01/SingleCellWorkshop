```python
A = np.array([[1,2],
              [2,4],
              [4,2]])
```

```python
fig, ax = plt.subplots(1, figsize=(5,5))

ax.scatter(A[:,0], A[:,1], s=100)

xlim = (0,5)
ylim = (0,5)

ax.set_xlim(xlim)
ax.set_ylim(ylim)



ax.set_xticks(np.arange(0, xlim[1] + 1))
ax.set_yticks(np.arange(0, ylim[1] + 1))

ax.text(1.1,1.7, '[1,2]')
ax.text(2.1,3.7, '[2,4]')
ax.text(4.1,1.7, '[4,2]')




fig.tight_layout()
fig.savefig('../img/2d_matrix_scatter.png', dpi=300)
```

```python
fig, ax = plt.subplots(1, figsize=(5,5))

ax.scatter(A[:,0], A[:,1], s=100)

xlim = (0,5)
ylim = (0,5)

ax.set_xlim(xlim)
ax.set_ylim(ylim)



ax.set_xticks(np.arange(0, xlim[1] + 1))
ax.set_yticks(np.arange(0, ylim[1] + 1))

ax.text(1.1,1.7, '[1,2]')
ax.text(2.1,3.7, '[2,4]')
ax.text(4.1,1.7, '[4,2]')

for i in [1,2,3,4]:
    ax.axvline(i, linestyle='--', color='lightgrey', zorder=0)
    ax.axhline(i, linestyle='--', color='lightgrey', zorder=0)


fig.tight_layout()
fig.savefig('../img/2d_matrix_scatter.lines.png', transparent=True, dpi=300)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10.5,5))

for i, ax in enumerate(axes):
    if i == 0:
        ax.set_title('Euclidean')
        ax.plot(A[[0,1],0], A[[0,1],1], linestyle='--', c='k', zorder=1)
        ax.plot(A[[0,2],0], A[[0,2],1], linestyle='--', c='k', zorder=1)

        dists = pdist(A, metric='Euclidean')
        ax.text(0.7, 2.75,   '{:.2f}'.format(dists[0]), color='darkred')
        ax.text(2.3, 2.1, '{:.2f}'.format(dists[1]), color='darkred')
    else:
        ax.set_title('Manhattan')
        ax.plot(A[[1,1],0], A[[0,1],1], linestyle='--', c='k', zorder=1)
        ax.plot(A[[0,2],0], A[[0,2],1], linestyle='--', c='k', zorder=1)

        dists = pdist(A, metric='cityblock')
        ax.text(1.3, 2.75, '{:.2f}'.format(dists[0]), color='darkred')
        ax.text(2.5, 2.1,  '{:.2f}'.format(dists[1]), color='darkred') 

        
    ax.scatter(A[:,0], A[:,1], zorder=2, s=100)

    ax.set_xlim([0,5])
    ax.set_ylim([0,5])
    
    ax.set_xticks(np.arange(0, xlim[1] + 1))
    ax.set_yticks(np.arange(0, ylim[1] + 1))
    
    ax.text(1.1,1.7, '[1,2]')
    ax.text(2.1,3.7, '[2,4]')
    ax.text(4.1,1.7, '[4,2]')


fig.tight_layout(w_pad=5)
fig.savefig('../img/2d_matrix_scatter.dists.png', dpi=300)
```
