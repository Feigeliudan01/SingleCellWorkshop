```python
import numpy as np
import matplotlib.pyplot as plt
import scprep
```

```python
means = [5, 12, 21]
stds = [2, 3, 6]
X = np.random.normal(means, stds, (100000, 3))
```

```python
colors = ['tab:blue', 'tab:purple', 'tab:green']
for i in range(X.shape[1]):
    ax = scprep.plot.histogram(X[:,i], xlabel='Gene Expression', range=(0, np.max(X)),
                          figsize=(6, 2.5), color=colors[i], cutoff=means[i])
    ax.annotate("", xytext=(means[i], 1000/stds[i]), xy=(means[i]+stds[i], 1000/stds[i]),
                arrowprops=dict(arrowstyle="->", color='tab:red', linewidth=2), fontsize=18)
    plt.tight_layout()
    plt.savefig("../img/sctransform/normal{}.png".format(i), dpi=300)
```

```python
means = [0,0,0]
stds = [1, 1, 1]
X = np.random.normal(means, stds, (100000, 3))
```

```python
colors = ['tab:blue', 'tab:purple', 'tab:green']
for i in range(X.shape[1]):
    ax = scprep.plot.histogram(X[:,i], xlabel='Gene Expression', range=(np.min(X), np.max(X)),
                          figsize=(6, 2.5), color=colors[i], cutoff=means[i])
    ax.annotate("", xytext=(means[i], 200/stds[i]), xy=(means[i]+stds[i], 200/stds[i]),
                arrowprops=dict(arrowstyle="->", color='tab:red', linewidth=2), fontsize=18)
    plt.tight_layout()
    plt.savefig("../img/sctransform/normal_rescaled{}.png".format(i), dpi=300)
```
